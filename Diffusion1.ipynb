```python
from random import randrange
from datetime import timedelta
import datetime

def random_date(start, end):
    """
    This function will return a random datetime between two datetime 
    objects.
    """
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)
```

```python
d1A = datetime.datetime.strptime('1/1/2016 1:30 PM', '%m/%d/%Y %I:%M %p')
d2A = datetime.datetime.strptime('3/1/2016 4:50 AM', '%m/%d/%Y %I:%M %p')
d1B = datetime.datetime.strptime('3/1/2017 1:30 PM', '%m/%d/%Y %I:%M %p')
d2B = datetime.datetime.strptime('5/1/2017 4:50 AM', '%m/%d/%Y %I:%M %p')
d1C = datetime.datetime.strptime('5/1/2018 1:30 PM', '%m/%d/%Y %I:%M %p')
d2C = datetime.datetime.strptime('8/1/2018 4:50 AM', '%m/%d/%Y %I:%M %p')
print(random_date(d1A, d2A))
```
